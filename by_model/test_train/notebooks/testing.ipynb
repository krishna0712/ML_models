```python
import pandas as pd
import seaborn as sns 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
iris = pd.read_csv("../data/iris_set",header = None)
iris.head()
```

```python
iris.rename(columns={0: "sepal_length", 1: "sepal_width", 2: "petal_length", 3: "petal_width", 4: "species"}, inplace=True)
iris.head()
```

```python
sns.scatterplot(x=iris['petal_length'], y=iris['sepal_length'], hue=iris['species'])
plt.title("Petal Length vs Sepal Length")
plt.xlabel("Petal Length")
plt.ylabel("Sepal Length")
plt.show()
```

```python
sns.scatterplot(x=iris['petal_length'], y=iris['petal_width'], hue=iris['species'])
plt.title("Petal Length vs Petal Width")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```
